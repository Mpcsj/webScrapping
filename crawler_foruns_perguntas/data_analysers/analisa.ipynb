```python
# descubrindo novas perguntas
find_all = lambda w,p: [x for x in range(w.find(p),len(w)) if  w[x:x + len(p)] == p]
limite = 80
dados = []
for col in df.columns:
    # para cada coluna
    for el in df[col]:
        #idx_perguntas = find_all(el,'?') # encontro todos os caracteres de interrogacao
        idx_perguntas = find_all(el,'como')
        for idx in idx_perguntas:
                dados.append(el[max(0,idx-50):min(idx+70,len(el))])
dados = [el.split('\n')[1] if '\n' in el else el for el in dados]
```

```python
dados = np.array(dados)
dados
```

```python
# retirando as perguntas ao estilo forum
perguntas_forum = ['alguém','alguem']
idx_perguntas = []
for i in range(len(dados)):
    for pergunta in perguntas_forum:
        if pergunta in dados[i]:
            idx_perguntas.append(i)
            break
dados = np.delete(dados,idx_perguntas)
print(len(idx_perguntas),' removidos')
del idx_perguntas
```
